# DDA

```python
start_point = [0,0]
end_point = [10,16]

#----- INITIALIZATION ------------------------
x1 = start_point[0]
y1 = start_point[1]
x2 = end_point[0]
y2 = end_point[1]

#----- CALCULATE DELTA --------------------------------
dx = x2 - x1
dy = y2 - y1

#----- CALCULATE STEP ---------------------------------
step = 0
if dx > dy:
    step = dx
else: step = dy

#----- CALCULATE INC  -----------------------------
x_inc = dx/step
y_inc = dy/step

#----- INITIALIZATION ARRAY -------------------------------
x = []
y = []

#----- GET INCREMENT X --------------------------
def increment_x(x1, x2, step):
    if x1 < x2:
        #print(x1)
        x.append(x1)
        return increment_x(x1 + step, x2, step)
    else:
        x.append(x2)
    
    return x

#----- GET INCREMENT Y --------------------------
def increment_y(y1, y2, step):
    if y1 < y2:
        #print(y1)
        y.append(y1)
        return increment_y(y1 + step, y2, step)
    else:
        y.append(y2)
    
    return y

#----- CALL FUNCTION OF INCREMENT --------------------------
increment_x(x1, x2, x_inc)
increment_y(y1, y2, y_inc)

print("x = ", x)
print("y = ", y)
#----- ROUND INCREMENT VALUE --------------------------
x = [round(item) for item in x]
y = [round(item) for item in y]

#----- DISPLAY ----------------------------
def display():
    
    print(
    f"""
    ===============================================================
                INITIALIZATION OF START POINT AND END POINT
    ===============================================================
    titik awal = {titik_awal}
    titik akhir = {titik_akhir}
    
    ===============================================================
                INITIALIZATION OF X AND Y
    ===============================================================
    x1 = {x1} | x2 = {x2}
    y1 = {y1} | y2 = {y2}
    
    ===============================================================
                GET DELTA  
    ===============================================================
    dx : {x2} - {x1} = {x2 - x1}
    dy : {y2} - {y1} = {y2 - y1}
    
    ===============================================================
                GET STEP
    ===============================================================
    dx : {dx}
    dy : {dy}
    
    if {dx} > {dy} = {dx}
    if {dx} < {dy} = {dy}
    
    step = {step}
    
    ===============================================================
                GET INC
    ===============================================================
    x_inc : dx/step = {dx/step}
    y_inc : dy/step = {dy/step}
    
    ===============================================================
                INCREMENT 
    ===============================================================
    x = {x}
    y = {y}

    """)
```

```python
import matplotlib.pyplot as plt
import numpy as np

display()
plt.figure(figsize=(15,8))
plt.xticks(np.arange(x[0], y[-1]+1, step=1))
plt.yticks(np.arange(x[0], y[-1]+1, step=1))
plt.scatter(x,y, color='r')
plt.plot(x,y, color= 'b')
plt.grid(True)
plt.title("Primitive Graphics")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
